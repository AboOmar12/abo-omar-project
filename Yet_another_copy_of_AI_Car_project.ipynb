Car 1 Arduino(Free) Code :


```python
#include <Wire.h>
#include <LiquidCrystal_I2C.h>
#include <SoftwareSerial.h>

// Initialize the LCD (address 0x27, 16 columns, 2 rows)
LiquidCrystal_I2C lcd(0x27, 16, 2);

// HC-05 Bluetooth pins
const int BT_RX = 10;  // Connect HC-05 TX to pin 10
const int BT_TX = 11;  // Connect HC-05 RX to pin 11
SoftwareSerial BTSerial(BT_RX, BT_TX);

// Pin definitions for the lights
const int lightPin1 = 8;  // First light (red)
const int lightPin2 = 9;  // Second light (blue)

float speed = 0;  // Variable to store speed from Bluetooth
const float speedMultiplier = 0.2677;  // Multiplier for speed calculation

// State variables
enum DisplayState { EMERGENCY, CAR_DETECTED, SHOW_SPEED };
DisplayState currentState = EMERGENCY;
unsigned long stateChangeTime = 0;

// Scrolling variables
int scrollPos = 0;
unsigned long lastScrollTime = 0;
const int scrollInterval = 150;  // Scroll speed (ms per character)
const int displayHoldTime = 0;  // Hold for 0 seconds after complete

// Timing variables
unsigned long lastFlashTime = 0;
const int flashInterval = 500;  // Flash every 500ms

void setup() {
  // Initialize Serial for debugging (optional)
  Serial.begin(9600);

  // Initialize Bluetooth Serial
  BTSerial.begin(9600);  // HC-05 default baud rate is 9600

  // Initialize the LCD
  lcd.init();
  lcd.backlight();  // Turn on the backlight

  // Set up the light pins as outputs
  pinMode(lightPin1, OUTPUT);
  pinMode(lightPin2, OUTPUT);

  // Display initial message
  currentState = EMERGENCY;
  stateChangeTime = millis();
  scrollPos = 0;
  lastScrollTime = millis();
}

void loop() {
  // Check if data is available from Bluetooth
  if (BTSerial.available()) {
    String receivedData = "";

    // Read incoming data until newline
    while (BTSerial.available()) {
      char c = BTSerial.read();
      if (c == '\n' || c == '\r') {
        break;
      }
      receivedData += c;
    }

    // Check if received data is 'w'
    if (receivedData.length() > 0) {
      // Print the exact number received
      Serial.print("Raw Data: ");
      Serial.println(receivedData);

      // If raw data is 'w', set speed to 53.54 cm/s
      if (receivedData == "w" || receivedData == "W") {
        speed = 53.54;
        Serial.print("Received 'w', speed set to: ");
        Serial.println(speed);
      }

      // Trigger the state sequence
      currentState = CAR_DETECTED;
      stateChangeTime = millis();
      scrollPos = 0;
      lastScrollTime = millis();
    }
  }

  // Handle state transitions based on timing
  if (currentState == CAR_DETECTED) {
    scrollText("Car Detected");
    if (millis() - stateChangeTime >= 5000) {  // 3 seconds max
      currentState = SHOW_SPEED;
      stateChangeTime = millis();
      scrollPos = 0;
      lastScrollTime = millis();
    }
  } else if (currentState == SHOW_SPEED) {
    scrollText("Speed: " + String(speed) + " cm/s");
    if (millis() - stateChangeTime >= 6000) {  // 3 seconds max
      currentState = EMERGENCY;
      stateChangeTime = millis();
      scrollPos = 0;
      lastScrollTime = millis();
    }
  } else if (currentState == EMERGENCY) {
    scrollText("EMERGENCY!");
    // Check if scroll is complete
    String displayText = "                EMERGENCY!                ";
    if (scrollPos >= displayText.length()) {
      // Reset for next loop
      scrollPos = 0;
      lastScrollTime = millis();
    }
  }

  // Handle the flashing lights
  flashAmbulanceLights();
}

void scrollText(String text) {
  unsigned long currentTime = millis();

  // Display "Ambulance Active" on top line
  lcd.setCursor(0, 0);
  lcd.print("Ambulance Active");

  // Check if it's time to scroll
  if (currentTime - lastScrollTime >= scrollInterval) {
    lastScrollTime = currentTime;

    // Billboard effect: text moves from right to left
    String displayText = "                " + text + "                ";  // Add padding

    if (scrollPos < displayText.length()) {
      lcd.setCursor(0, 1);
      // Display 16 characters at current scroll position
      lcd.print(displayText.substring(scrollPos, scrollPos + 16));
      scrollPos++;
    } else {
      // Scroll complete, hold for 3 seconds then reset
      if (currentTime - lastScrollTime >= displayHoldTime) {
        scrollPos = 0;
      }
    }
  }
}

void flashAmbulanceLights() {
  unsigned long currentTime = millis();

  // Check if it's time to flash
  if (currentTime - lastFlashTime >= flashInterval) {
    lastFlashTime = currentTime;

    // Alternate the lights
    static boolean state = false;
    state = !state;

    if (state) {
      digitalWrite(lightPin1, HIGH);
      digitalWrite(lightPin2, LOW);
    } else {
      digitalWrite(lightPin1, LOW);
      digitalWrite(lightPin2, HIGH);
    }
  }
}
```

Car 2 PIC-Code :

```python
#include <xc.h>
#include <stdint.h>

// Oscillator Frequency
#define _XTAL_FREQ 4000000
// ═══════════════════════════════════════════════════════════════════════
// PIN DEFINITIONS
// ═══════════════════════════════════════════════════════════════════════

// HC-SR04 Ultrasonic Sensor Pins
#define TRIG_PIN PORTBbits.RB0
#define TRIG_TRIS TRISBbits.TRISB0
#define ECHO_PIN PORTBbits.RB1
#define ECHO_TRIS TRISBbits.TRISB1

// LED Pins (Visual Indicators)
#define LEFT_LED PORTBbits.RB2
#define LEFT_LED_TRIS TRISBbits.TRISB2
#define RIGHT_LED PORTBbits.RB3
#define RIGHT_LED_TRIS TRISBbits.TRISB3

// Digital Signal Pin to Arduino (HIGH/LOW communication)
#define SIGNAL_PIN PORTCbits.RC6
#define SIGNAL_TRIS TRISCbits.TRISC6

// ═══════════════════════════════════════════════════════════════════════
// CONFIGURATION PARAMETERS
// ═══════════════════════════════════════════════════════════════════════

#define DISTANCE_THRESHOLD 35   // Detection threshold in centimeters
#define BLINK_COUNT 5           // Number of LED blinks per sequence
#define BLINK_DURATION 500      // Duration of each blink state (ms)
#define SEQUENCE_DELAY 500      // Delay between blink sequences (ms)

// ═══════════════════════════════════════════════════════════════════════
// FUNCTION PROTOTYPES
// ═══════════════════════════════════════════════════════════════════════

void System_Init(void);
void Timer1_Init(void);
void Trigger_Pulse(void);
uint16_t Get_Distance(void);
void Blink_Left_LED(uint8_t times, uint16_t duration);

// ═══════════════════════════════════════════════════════════════════════
// MAIN PROGRAM
// ═══════════════════════════════════════════════════════════════════════

void main(void) {
    // Initialize all systems
    System_Init();

    // System ready indicator: Quick blink left LED 3 times
    for(uint8_t i = 0; i < 3; i++) {
        LEFT_LED = 1;
        __delay_ms(150);
        LEFT_LED = 0;
        __delay_ms(150);
    }
    __delay_ms(500);

    // Flag to execute LED sequence only once per detection
    uint8_t sequence_executed = 0;

    // Main loop - runs forever
    while(1) {
        // Measure distance using HC-SR04
        uint16_t distance = Get_Distance();

        // Check if obstacle detected within threshold
        if(distance < DISTANCE_THRESHOLD && distance > 0) {

            // Execute LED sequence only once when first detected
            if(!sequence_executed) {
                // IMMEDIATELY send HIGH signal to Arduino
                SIGNAL_PIN = 1;

                // SEQUENCE STEP 1: Blink left LED
                Blink_Left_LED(BLINK_COUNT, BLINK_DURATION);

                // SEQUENCE STEP 2: Delay between sequences
                __delay_ms(SEQUENCE_DELAY);

                // SEQUENCE STEP 3: Blink left LED again
                Blink_Left_LED(BLINK_COUNT, BLINK_DURATION);

                // Turn off all LEDs after sequence
                LEFT_LED = 0;
                RIGHT_LED = 0;

                // Mark sequence as executed
                sequence_executed = 1;
            }

            // Keep signal HIGH while obstacle is present
            SIGNAL_PIN = 1;
            __delay_ms(100);
        }
        else {
            // No obstacle detected - safe distance

            // Send LOW signal to Arduino
            SIGNAL_PIN = 0;

            // Turn off all LEDs
            LEFT_LED = 0;
            RIGHT_LED = 0;

            // Reset sequence flag for next detection
            sequence_executed = 0;

            __delay_ms(100);
        }
    }
}

// ═══════════════════════════════════════════════════════════════════════
// INITIALIZATION FUNCTIONS
// ═══════════════════════════════════════════════════════════════════════

void System_Init(void) {
    // Initialize HC-SR04 pins
    TRIG_TRIS = 0;      // TRIG as output
    ECHO_TRIS = 1;      // ECHO as input
    TRIG_PIN = 0;       // TRIG starts LOW

    // Initialize LED pins
    LEFT_LED_TRIS = 0;  // Left LED as output
    LEFT_LED = 0;       // Start OFF
    RIGHT_LED_TRIS = 0; // Right LED as output
    RIGHT_LED = 0;      // Start OFF

    // Initialize signal pin
    SIGNAL_TRIS = 0;    // Signal pin as output
    SIGNAL_PIN = 0;     // Start LOW (no obstacle)

    // Initialize Timer1 for distance measurement
    Timer1_Init();
}

void Timer1_Init(void) {
    T1CON = 0x00;       // Timer1 OFF, 1:1 prescaler, internal clock
    TMR1H = 0;          // Clear Timer1 high byte
    TMR1L = 0;          // Clear Timer1 low byte
}

// ═══════════════════════════════════════════════════════════════════════
// HC-SR04 ULTRASONIC SENSOR FUNCTIONS
// ═══════════════════════════════════════════════════════════════════════

void Trigger_Pulse(void) {
    // Send 10us trigger pulse to HC-SR04
    TRIG_PIN = 1;
    __delay_us(10);
    TRIG_PIN = 0;
}

uint16_t Get_Distance(void) {
    uint16_t timer_count;
    uint32_t distance_cm;

    // Send trigger pulse
    Trigger_Pulse();

    // Wait for ECHO pin to go HIGH (with timeout)
    uint16_t timeout = 5000;
    while(!ECHO_PIN && timeout--) {
        __delay_us(1);
    }
    if(timeout == 0) return 9999;  // Timeout - no echo received

    // Start Timer1 to measure pulse width
    TMR1H = 0;
    TMR1L = 0;
    T1CONbits.TMR1ON = 1;

    // Wait for ECHO pin to go LOW (with timeout)
    timeout = 30000;
    while(ECHO_PIN && timeout--) {
        // Check for Timer1 overflow
        if(TMR1H > 0xF0) break;
    }

    // Stop Timer1
    T1CONbits.TMR1ON = 0;

    // Read Timer1 value
    timer_count = (TMR1H << 8) | TMR1L;

    /*
     * Distance Calculation:
     * ────────────────────
     * At 4MHz, Timer1 increments every 1μs (instruction cycle = Fosc/4)
     *
     * Distance formula:
     *   Distance (cm) = (Time × Speed_of_Sound) / 2
     *   Distance (cm) = (timer_count × 1μs × 34300 cm/s) / 2
     *   Distance (cm) = (timer_count × 0.0343) / 2
     *   Distance (cm) = timer_count / 58.24
     *
     * Using integer division: distance = timer_count / 58
     */
    distance_cm = timer_count / 58;

    // Return distance (limited to reasonable range)
    if(distance_cm > 400) return 9999;  // Out of range

    return (uint16_t)distance_cm;
}

// ═══════════════════════════════════════════════════════════════════════
// LED BLINKING FUNCTIONS
// ═══════════════════════════════════════════════════════════════════════
void Blink_Left_LED(uint8_t times, uint16_t duration) {
    /*
     * Blinks left LED only
     * Used for turn signal / directional indicator
     *
     * Parameters:
     *   times    - Number of complete blink cycles
     *   duration - Duration in milliseconds for ON and OFF states
     */
    for(uint8_t i = 0; i < times; i++) {
        // Turn left LED ON
        LEFT_LED = 1;

        // Stay ON for specified duration
        for(uint16_t j = 0; j < duration; j++) {
            __delay_ms(1);
        }

        // Turn left LED OFF
        LEFT_LED = 0;

        // Stay OFF for specified duration
        for(uint16_t j = 0; j < duration; j++) {
            __delay_ms(1);
        }
    }
}
```
